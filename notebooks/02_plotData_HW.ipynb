# Explore relationship between data and label

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
sns.set()
%matplotlib notebook

data_path = os. getcwd()[:-9] +'data/'
```

```python
df =  pd.read_csv(data_path+'df_mvp.csv')
df.time_bin = pd.to_datetime(df.time_bin)
```

```python
safe_pop= df[df['label']==0]['current_popularity']
danger_pop = df[df['label']==1]['current_popularity']

sns.distplot( safe_pop, color="lightseagreen", label="safe")
sns.distplot( danger_pop, color="tomato", label="danger")
plt.title('Large overlap between "safe" & "danger" in the Google populartimes data')
plt.legend()
plt.show()
```

```python
from scipy import stats
stats.ttest_ind(safe_pop, danger_pop)
# Although the two distributions are different, they have a large overlap
# meaning that a given current_popularity value cannot fully determine how busy the park is. 
```

```python
plt.figure()
safe_temp= df[df['label']==0]['temp']
danger_temp = df[df['label']==1]['temp']

sns.distplot( safe_temp, color="lightseagreen", label="safe")
sns.distplot( danger_temp, color="tomato", label="danger")
plt.title('Large overlap between "safe" & "danger" in temperature')
plt.legend()
plt.show()
```

```python
stats.ttest_ind(safe_temp, danger_temp)
```

```python
plt.figure()
safe_wind= df[df['label']==0]['wind_speed']
danger_wind = df[df['label']==1]['wind_speed']

sns.distplot( safe_wind, color="lightseagreen", label="safe")
sns.distplot( danger_wind, color="tomato", label="danger")
plt.title('Large overlap between "safe" & "danger" in wind_speed')
plt.legend()
plt.show()
```

```python
stats.ttest_ind(safe_wind, danger_wind)
```
